```python
f = open("./command_data.csv", "r", encoding="utf8")
data = f.read()
f.close()
```

```python
command_dict = {}

for line in data.split("\n"):
    lines = line.split(",")
    if len(lines) != 2:
        continue
    command, key = line.split(",")
    if key in command_dict.keys():
        command_dict[key] += 1
    else:
        command_dict[key] = 1
```

```python
def get_key(item):
    return item[1]
```

```python
dictlist = []
for key, value in command_dict.items():
    dictlist.append([key, value])

sorted(dictlist, key=get_key, reverse=True)
```
